```python
from IPython.display import display
import nltk
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cmx
import matplotlib.colors as colors
import pandas as pd
from collections import Counter
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/fahmida/Desktop/Rental/airbnb/listings_clean.csv')
```

```python
zip_counts = Counter(df.zipcode)
zips = pd.DataFrame.from_dict(zip_counts, orient='index').sort_values(by=0)
zips['z_price'] = 0

for index,row in zips.iterrows():
    df_zip_price = df[df.zipcode == index]
    zip_price = df_zip_price.price.mean()
    zips.loc[index,'z_price'] = zip_price
    
highest_zip = zips.tail(25)
highest_zip[0].plot(kind='bar',figsize=(20,20),title='Total Listings by Zip Code and Price')
```

```python
nb_counts = Counter(df.neighbourhood)
neighbs = pd.DataFrame.from_dict(nb_counts, orient='index').sort_values(by=0)
neighbs['n_price'] = 0

for index,row in neighbs.iterrows():
    df_neigh_price = df[df.neighbourhood== index]
    neigh_price = df_neigh_price.price.mean()
    neighbs.loc[index,'n_price'] = zip_price
    
highest_neighbs = neighbs.tail(25)
highest_neighbs[0].plot(kind='pie',figsize=(15,15),title='Total Listings by Neighborhood and Price')
```
