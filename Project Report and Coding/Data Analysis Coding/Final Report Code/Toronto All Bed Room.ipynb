```python
#### Load all types of toronto rooms
# import all modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
matplotlib.use('agg')
import matplotlib.pyplot as plt
%matplotlib inline
fig1 = plt.figure(figsize=(2, 2), dpi=100)
%config InlineBackend.figure_format = 'retina'
import datetime
from matplotlib.dates import DateFormatter
from matplotlib.dates import HourLocator
# Import Multiple CSVs and combining them
Total_filepath = "C:/Users/fahmida/Desktop/Rental/Permanent Rental/By Times/Primary Rental Comparision(Apartment)/Summary By Room Type/Total Apartment Toronto_cleaned.csv"
df= pd.read_csv(Total_filepath, header=0, index_col=None) 
print(df)
# Print the summary statistics of the DataFrame
print(df.describe())
```

```python
#bytime = df.groupby("Time")
#bytime["Vacancy Rate (%)"].describe()
#bytime["Vacancy Rate (%)"].mean()

# Convert the date column to a datestamp type
df['Year'] = pd.to_datetime(df['Year'])

# Set the date column as the index of your DataFrame meat
df = df.set_index('Year')
print(df)
# Print the summary statistics of the DataFrame
print(df.describe())
```

```python
# Imputing Missing values
# Convert '?' to NaN
df[df == 'NA'] = np.nan

# Print the number of NaNs
print(df.isnull().sum())

# Print shape of original DataFrame
print("Shape of Original DataFrame: {}".format(df.shape))
# fill missing values with mean column values
df.fillna(df.mean(), inplace=True)
# count the number of NaN values in each column
print(df.isnull().sum())
#print("Shape of DataFrame After Filling up with Average Column Value
#print(df.head(100))
```

```python
# Plot Vacancy Rate For All Bed Room Types dataset
df1=df.drop(["Availability Rate (%)_Total Apartment","Average Rent ($)_ Total Apartment","Median Rent ($)_Total Apartment", "% Change_Total Apartment", "Units_Total Apartment", "Availability Rate (%)_Bachelor",   
"Average Rent ($)_Bachelor", "Median Rent ($)_bachelor", "% Change_Bachelor", "Units_Bachelor","Availability Rate (%)_1 Bed", "Average Rent ($)_1 Bed","Median Rent ($)_1 Bed","% Change_ 1 Bed", "Units_ 1 Bed","Availability Rate_2 Bed (%)", 
"Average Rent_2 Bed ($)","Median Rent_2 Bed ($)","Units_2 Bed","% Change_2 Bed","Availability Rate (%)_3+ Bed", "Average Rent ($)_3+ Bed","Median Rent ($)_3+ Bed","% Change_3+ Bed","Units_3+ Bed"], axis=1)
ax = df1.plot(linewidth=2, fontsize=8)

# Additional customizations
ax.set_xlabel('Date')
ax.legend(fontsize=8)

# Show plot
plt.show()
```

```python
# Plot an area chart
ax = df1.plot.area(fontsize=8)

# Additional customizations
ax.set_xlabel('Date')
ax.legend(fontsize=8)

# Show plot
plt.show()
```

```python
# Import seaborn library
import seaborn as sns

# Get correlation matrix of the meat DataFrame
corr_df1 = df1.corr(method='spearman')

# Customize the heatmap of the corr_meat correlation matrix
sns.heatmap(corr_df1,
            annot=True,
            linewidths=0.4,
            annot_kws={"size": 10})

plt.xticks(rotation=90)
plt.yticks(rotation=0) 
plt.show()
```

```python
# Clustered heatmaps
# Import seaborn library
import seaborn as sns

# Get correlation matrix of the DataFrame
corr_df1 = df1.corr(method='pearson')

# Customize the heatmap of the corr_meat correlation matrix and rotate the x-axis labels
fig = sns.clustermap(corr_df1,
                     row_cluster=True,
                     col_cluster=True,
                     figsize=(10, 10))

plt.setp(fig.ax_heatmap.xaxis.get_majorticklabels(), rotation=90)
plt.setp(fig.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.show()
```
